```python
import pandas as pd
import os
from glob import glob
```

```python
DATA_PATH = 'C:/Users/Andy/Documents/guo_pengcheng_001643483/Python4DataAnalysis/DataAnalysisPython/final _project/Data/Crime Incident Reports/Crime Incident Reports.csv'
data = pd.read_csv(DATA_PATH)
del data['Unnamed: 0']
data
```

```python
typelist = []
for x in data['incident_type_description']:
    if x not in typelist:
        typelist.append(x)
print(typelist)
len(typelist)
```

```python
datelist = []
for x in data['fromdate']:
    if x.split('T')[0] not in datelist:
        datelist.append(x.split('T')[0])
```

```python
print(datelist)
len(datelist)
```

```python
PATH_TO_DATA = 'C:/Users/Andy/Documents/guo_pengcheng_001643483/Python4DataAnalysis/DataAnalysisPython/final _project/Data/Crime Incident Reports/data'
for x in datelist:
    if not os.path.exists(PATH_TO_DATA + '/' + x):
        os.makedirs(PATH_TO_DATA + '/' + x)
    for y in typelist:
        if not os.path.exists(PATH_TO_DATA + '/'+ x + '/' + y):
            os.makedirs(PATH_TO_DATA + '/'+ x + '/' + y)
len(typelist)
```

```python
for x in typelist:
    for y in datelist:
        df1 = data.loc[data['fromdate'].str.contains(y)]
        df2 = df1.loc[df1['incident_type_description'].str.contains(x)]
        path = PATH_TO_DATA + '/' + y + '/' + x + '/' + 'data.csv'
        df2.to_csv(path)
```
