```python
import pandas as pd
import os
from glob import glob
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA_PATH = 'C:/Users/Andy/Documents/guo_pengcheng_001643483/Python4DataAnalysis/DataAnalysisPython/final _project/Analysis'
```

```python
data1 = pd.read_csv(DATA_PATH + '/Analysis2/Average_Earning.csv')
del data1['Unnamed: 0']
data1
```

```python
data2 = pd.read_csv(DATA_PATH + '/Analysis3/Average_Incident.csv')
del data2['Unnamed: 0']
data2
```

```python
data1['Type'] = 'Earning'
data1.columns = ['Year','Amount','Type']
data1
```

```python
data2['Type'] = 'Incident'
data2.columns = ['Year','Amount','Type']
data2
```

```python
data3 = data1.append(data2,ignore_index=True)
data3
```

```python
plt.subplots(figsize=(20,8))     
sns.set_style("whitegrid")
colors = ['#624ea7', 'g']
ax = sns.barplot(x="Year", y="Amount",hue="Type", data=data3, palette=colors)  
ax.set_title('Number Of Incident Each Year')     
plt.setp(ax.xaxis.get_majorticklabels(), rotation=30)       
ax.xaxis.get_label().set_fontsize(30)      
ax.yaxis.get_label().set_fontsize(30)
ax.title.set_fontsize(45)                  
ax.tick_params(axis='x', which='major', labelsize=24)      
ax.tick_params(axis='y', which='major', labelsize=24)
plt.savefig("output1.png")
```
