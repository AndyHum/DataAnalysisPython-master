```python
import pandas as pd
import os
from glob import glob
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATA_PATH = 'C:/Users/Andy/Documents/guo_pengcheng_001643483/Python4DataAnalysis/DataAnalysisPython/final _project/Data/Crime Incident Reports/Crime Incident Reports.csv'
data = pd.read_csv(DATA_PATH)
del data['Unnamed: 0']
data
```

```python
count2012 = 0
for x in data['fromdate']:
     if x.split('-')[0] == '2012':
            count2012 = count2012 + 1
print(count2012)
```

```python
count2013 = 0
for x in data['fromdate']:
     if x.split('-')[0] == '2013':
            count2013 = count2013 + 1
print(count2013)
```

```python
count2014 = 0
for x in data['fromdate']:
     if x.split('-')[0] == '2014':
            count2014 = count2014 + 1
print(count2014)
```

```python
count2015 = 0
for x in data['fromdate']:
     if x.split('-')[0] == '2015':
            count2015 = count2015 + 1
print(count2015)
```

```python
s1 = pd.Series(['2012','2013','2014','2015'], name='Year')
s2 = pd.Series([count2012,count2013,count2014,count2015], name='Number Of Incident')
df = pd.concat([s1,s2],axis=1)
df
PATH_TO_DATA = 'C:/Users/Andy/Documents/guo_pengcheng_001643483/Python4DataAnalysis/DataAnalysisPython/final _project/Analysis'
path5 = PATH_TO_DATA + '/' + 'Analysis3/' + 'Average_Incident.csv'
df.to_csv(path5)
```

```python
plt.subplots(figsize=(20,8))     
sns.set_style("whitegrid")
colors = ['#624ea7', 'g', 'yellow', 'k']
ax = sns.barplot(x="Year", y="Number Of Incident", data=df, palette=colors)  
ax.set_title('Number Of Incident Each Year')     
plt.setp(ax.xaxis.get_majorticklabels(), rotation=30)       
ax.xaxis.get_label().set_fontsize(30)      
ax.yaxis.get_label().set_fontsize(30)
ax.title.set_fontsize(45)                  
ax.tick_params(axis='x', which='major', labelsize=24)      
ax.tick_params(axis='y', which='major', labelsize=24)
plt.savefig("output.png")
```
