```python
import requests
import json
import csv
from pandas.io.json import json_normalize
import os
from glob import glob
import urllib
```

```python
url1 = 'https://data.cityofboston.gov/resource/ufcx-3fdn.json?$limit=300000'
r1 = requests.get(url1)
r1.json()
```

```python
PATH_TO_DATA = 'Data'
df1 = json_normalize(r1.json())
if not os.path.exists(PATH_TO_DATA + '/' + 'Crime Incident Reports'):
    os.makedirs(PATH_TO_DATA + '/' + 'Crime Incident Reports')
path1 = PATH_TO_DATA + '/' + 'Crime Incident Reports/' + 'Crime Incident Reports.csv'
column = list(df1.columns.values)
df1.to_csv(path1)
df1
```
